```python
df.drop_duplicates(inplace=True)
```

```python
# Fill logically based on type
df.loc[df['type'] == 'Movie', 'duration_seasons'] = 0
df.loc[df['type'] == 'TV Show', 'duration_minutes'] = 0
```

```python
df.dropna(subset=['type', 'title'], inplace=True)
```

```python
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```

```python
df['release_year'] = df['release_year'].fillna(df['release_year'].median())
```

```python
df['rating'] = df['rating'].fillna('Not Rated')
```

```python
df['description'] = df['description'].fillna('No Description Available')
```

```python
df.dropna(subset=['show_id'], inplace=True)
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df['show_id'] = df['show_id'].astype(int).astype(str)
```

```python
df['content_age'] = 2026 - df['release_year']
```

```python
df['content_type_flag'] = df['type'].map({'Movie': 1, 'TV Show': 0})
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python

df['type'] = df['type'].str.strip()
```

```python
df['is_movie'] = df['type'].apply(lambda x: 1 if x == 'Movie' else 0)
```

```python
df.to_csv(r"C:\Users\MY-PC\Desktop\cleaned_netflix_data.csv", index=False)
```
